```python
# build_nutrition_db.py
import pandas as pd
import sqlite3
```

```python
# 1️⃣ Load the CSV file
df = pd.read_csv("nutrition.csv")
```

```python
# 2️⃣ Create SQLite DB and store the data
conn = sqlite3.connect("nutrition_local.db")
df.to_sql("foods", conn, if_exists="replace", index=False)
conn.close()

print("✅ Local nutrition database created successfully!")
```
